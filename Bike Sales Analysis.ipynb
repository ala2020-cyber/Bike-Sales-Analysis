## Cleaning Data

```python
# remove all duplicated rows
sales.drop_duplicates(inplace = True)
```

```python
sales.info()
```

## Correlation

```python
sales.corr().style.background_gradient(cmap='coolwarm')
```

## How many product ?


```python
nb_products=sales.Product.value_counts().count()
nb_products
```

## Revenue per product


```python
revenue_per_product=sales.groupby("Product").sum().filter(["Revenue"])
# or
# c=sales.groupby("Product").sum().filter(["Revenue"]).loc["Classic Vest, M"]
revenue_per_product
```

## Best selling product


```python
Best_selling_product=sales.groupby("Product").sum().sort_values('Revenue',ascending=False).filter(["Revenue"])
Best_selling_product.head(1)
```

## The 3 higther revenue by state


```python
sales.groupby("State").sum().sort_values("Revenue",ascending=False).filter(["Revenue"]).head(3)
```

## In which month was the higther profit and revenue ?


```python
sales.groupby(by=["Month","Year"]).sum().sort_values("Profit",ascending=False).filter(["Profit","Revenue"]).head(1)
```
